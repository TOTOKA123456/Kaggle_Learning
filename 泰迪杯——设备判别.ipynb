```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = open('D:/Data/泰迪杯/谐波电压/YD1_谐波数据_UC.csv', encoding = 'utf-8')
YD1 = pd.read_csv(f)
YD1.head()
```

```python
x = range(len(YD1))
plt.plot(x, YD1['0'])
plt.plot(x, YD1['10'])
plt.plot(x, YD1['20'])
plt.plot(x, YD1['30'])
plt.plot(x, YD1['40'])
plt.plot(x, YD1['50'])
plt.plot(x, YD1['60'])
plt.plot(x, YD1['70'])
plt.plot(x, YD1['80'])
plt.plot(x, YD1['90'])
plt.plot(x, YD1['100'])
plt.plot(x, YD1['120'])
plt.plot(x, YD1['140'])
plt.plot(x, YD1['160'])
plt.plot(x, YD1['188'])
```

上图中，不同颜色代表三个处于不同运行状态下的YD1设备的谐波电压波形，非常相似。
